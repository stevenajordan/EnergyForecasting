```python
from math import sqrt

#numpy 
import numpy as np
from numpy import concatenate

#pandas
import pandas as pd
pd.set_option('display.max_rows', 365)
pd.set_option('display.max_columns', 365)
from pandas import to_datetime
from pandas import read_csv
from pandas import DataFrame
from pandas import concat

#matplotlib
from matplotlib import pyplot

#scipy
from scipy import stats

#sklearn
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso

from sklearn.model_selection import train_test_split,ShuffleSplit
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
scenario_daily = pd.read_csv('scenario_daily_agg2.csv', index_col=False)
scenario_daily.rename(columns={'Wind Speed_AVG':'Wind_Speed_AVG', 'Solar_Elevation_GEN_AVG':'Solar_Elevation_AVG'}, inplace=True)

product_solar_daily = pd.read_csv('solar_daily_agg.csv', index_col=False)
product_wind_daily = pd.read_csv('wind_daily_agg.csv', index_col=False)
consump_food_daily = pd.read_csv('consump_fs_daily_agg.csv', index_col=False)
consump_groc_daily = pd.read_csv('consump_g_daily_agg.csv', index_col=False)
consump_heal_daily = pd.read_csv('consump_hc_daily_agg.csv', index_col=False)
consump_k12_daily = pd.read_csv('consump_k12_daily_agg.csv', index_col=False)
consump_lodg_daily = pd.read_csv('consump_l_daily_agg.csv', index_col=False)
consump_off_daily = pd.read_csv('consump_o_daily_agg.csv', index_col=False)
consump_res_daily = pd.read_csv('consump_r_daily_agg.csv', index_col=False)
consump_sar_daily = pd.read_csv('consump_sar_daily_agg.csv', index_col=False)
```

```python
rand = 3
#split 10 datasets into train and test sets, with train/test proportions of 80/20
solar_train, solar_test = train_test_split(product_solar_daily, test_size=0.2, random_state=rand)
wind_train, wind_test = train_test_split(product_wind_daily, test_size=0.2, random_state=rand)
food_train, food_test = train_test_split(consump_food_daily, test_size=0.2, random_state=rand)
groc_train, groc_test = train_test_split(consump_groc_daily, test_size=0.2, random_state=rand)
heal_train, heal_test = train_test_split(consump_heal_daily, test_size=0.2, random_state=rand)
k12_train, k12_test = train_test_split(consump_k12_daily, test_size=0.2, random_state=rand)
lodg_train, lodg_test = train_test_split(consump_lodg_daily, test_size=0.2, random_state=rand)
off_train, off_test = train_test_split(consump_off_daily, test_size=0.2, random_state=rand)
res_train, res_test = train_test_split(consump_res_daily, test_size=0.2, random_state=rand)
sar_train, sar_test = train_test_split(consump_sar_daily, test_size=0.2, random_state=rand)
```

```python
#Wind Lasso
wind_x = wind_train.drop(columns = ['Electricity_KW_HR_AVG', 'Date_Time'])
wind_y = wind_train['Electricity_KW_HR_AVG']

estimator = Lasso()
featureSelection = SelectFromModel(estimator)
featureSelection.fit(wind_x, wind_y)

selectedFeatures = featureSelection.transform(wind_x)
#selectedFeatures

fs = wind_x.columns[featureSelection.get_support()]
#remaining features
keep = ['Electricity_KW_HR_AVG']
for i in range(len(fs)):
    keep.append(fs[i])
#keep

wind_train_reduced = wind_train[keep]
wind_test_reduced = wind_test[keep]
wind_scenario_reduced = scenario_daily[keep[1:]]
```
